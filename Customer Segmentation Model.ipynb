##Choosing annual income column and Spending Score Column

```python
x = dataset.iloc[:,[3,4]].values
print(x)
```

##Choosing number of Clusters
- We use WCSS Method (Within Cluster Sum Of Squares)

```python
import sklearn.cluster as skcluster
import seaborn as sns
import matplotlib.pyplot as plt # Import matplotlib

# Assuming 'x' is defined somewhere with your data
wcss = []
for i in range(1, 11):
    kmean = skcluster.KMeans(n_clusters=i, init='k-means++', random_state=63)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss) # 'wcss' is now populated after the loop
plt.title('The Elbow PLot of WCSS')
plt.xlabel('Numbers of Clusters')
plt.ylabel('WCSS')
plt.show()
```

##Optimal no. of clusters = 5
Training K-Mean Clustering Model

```python
kmean = skcluster.KMeans(n_clusters=5, init='k-means++', random_state=23)
y = kmean.fit_predict(x)
print(y)
```

##Visualizing the Data

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='blue',label = 'Cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='red',label = 'Cluster 2')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='green',label = 'Cluster 3')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='purple',label = 'Cluster 4')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='orange',label = 'Cluster 5')

plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s=100,c='cyan',label = 'Centroids')

plt.title('Clusters of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending of Scores')
plt.show()
```
